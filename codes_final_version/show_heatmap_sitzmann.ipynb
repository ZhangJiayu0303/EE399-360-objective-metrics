```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import cv2

font_t = {'weight' : 'bold'}

# Read votes

# vd = pd.read_csv('_votesData.csv',index_col=0,dtype={'SubjectID':str,'W':int,'H':int,'Q':str}).reset_index(drop=True)
# vd.head()
```

```python
fig = plt.figure(figsize=(14,21))
count = 1
heat = np.load('/Users/zhangjiayu/Desktop/sal_heatmaps.npy', encoding = "latin1").item()
heatmaps = {}

list_name = ['07', '13', '15', '04', '18', '02', '19', '06', '16', '12']
image_name = ['Clock Tower', 'Snow', 'Shark', 'Swimming Pool', 'Railway', 'Seashore', \
              'Cabin', 'Chess', 'Office', 'Bedroom']
for i,f in enumerate(list_name):
    heatmaps['sitzmann00' + list_name[i]] = cv2.resize(heat['sitzmann_00' + list_name[i]], \
                                                      dsize=(8192, 4096), interpolation=cv2.INTER_CUBIC) 
    ax = fig.add_subplot(5, 2, i+1)
    #img = plt.imread('../images/'+f)
    ax.imshow(heatmaps['sitzmann00' + list_name[i]])
    name = image_name[i]
    plt.title(name, **font_t)
plt.show()
fig.savefig("/Users/zhangjiayu/Desktop/sitzmann_heatmap.png", dpi = 600)
```
